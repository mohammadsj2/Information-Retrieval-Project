```python
# setup preprocessor
preprocessor.set_params(
    vectorizer__n_components=256, # dimensionality of the word vectors
    vectorizer__w2v={
        'iter':128, # number of training iterations of the w2v model
        'min_count':2, # all words with total frequency lower than this will be ignored
    },
)
preprocessor.fit(train_data)
train_data = preprocessor.transform(train_data)
test_data = preprocessor.transform(test_data)

train_data.head(2) # see how the preprocessed data looks!
```

   هم برای اعتبار‌سنجی عمل‌کرد الگوریتم‌های یادگیری ماشین و انتخاب بهینه‌تر پارامتر‌های آنها وجود دارد و به جای این‌کار می‌توانید از آنها استفاده‌کنید.

```python
# optional
np.random.seed(546)
validation_fraction = 0.1 # validation proportion
mask = np.random.rand(len(train_data)) < (1-validation_fraction)
val_split = train_data.iloc[~mask]
train_split = train_data.iloc[mask]
```

```python
def get_vec1(table):
    return table['vec_1'].values.tolist()

def get_vec2(table):
    tmp = table['vec_2'].values.tolist()
    return [x.tolist() for x in tmp]

val_split_vec1=get_vec1(val_split)
val_split_vec2=get_vec2(val_split)
train_split_vec1=get_vec1(train_split)
train_split_vec2=get_vec2(train_split)
```

```python
def get_pop(table):
    return table['popularity'].values.tolist()
train_split_pop = get_pop(train_split)
val_split_pop = get_pop(val_split)
```

```python
# [bonus] visualize clusters
def draw_clustering(title,test_vec,pred):
    df = pd.DataFrame(data=[[x for x in test_vec[i]] for i in range(len(test_data))])
    pca=PCA(n_components=2)
    tmp = pca.fit_transform(df)
    principalDf = pd.DataFrame(data = tmp, columns = ['principal component 1', 'principal component 2'])
    df=pd.concat([principalDf, pd.DataFrame(data=[[pred[i]] for i in range(len(test_data))],columns=['target'])], axis = 1)
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('Principal Component 1', fontsize = 15)
    ax.set_ylabel('Principal Component 2', fontsize = 15)
    ax.set_title(title, fontsize = 20)


    targets = [i for i in range(16)]
    names = ['class '+str(i) for i in range(16)]
    colors = ['#641E16','#512E5F','#154360','#17A589','#82E0AA','#F4D03F','#E67E22','#7B7D7D','#E6B0AA','#5D6D7E','#212F3D','#FDEBD0','#D2B4DE','#AAB7B8','#D4E6F1','#FC00F0']
    for target, color in zip(targets,colors):
        indicesToKeep = df['target'] == target
        ax.scatter(df.loc[indicesToKeep, 'principal component 1']
                   , df.loc[indicesToKeep, 'principal component 2']
                   , c = color
                   , s = 7)
    ax.legend(names)
    ax.grid()
draw_clustering('actual clustering by genre vec_1',test_vec1,test_genre)
draw_clustering('actual clustering by genre vec_2',test_vec2,test_genre)

draw_clustering('KMeans clustering by genre vec_1',test_vec1,pred1KM)
draw_clustering('KMeans clustering by genre vec_2',test_vec2,pred2KM)

draw_clustering('GMM clustering by genre vec_1',test_vec1,pred1GMM)
draw_clustering('GMM clustering by genre vec_2',test_vec2,pred2GMM)

draw_clustering('HC clustering by genre vec_1',test_vec1,pred1HC)
draw_clustering('HC clustering by genre vec_2',test_vec2,pred2HC)
```
