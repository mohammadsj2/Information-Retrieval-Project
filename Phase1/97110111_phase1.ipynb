```python
import nltk
import string
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
import csv

# man az in link baraye tabe komak gereftam: https://tomelf.github.io/nlp/machine%20learning/text-preprocessing/
# va az in link baraye khoondane file .csv: https://realpython.com/python-csv/

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

all_original_words={}

def tokenize(raw_text, add_to_words=True):
    global all_original_words
    raw_text = raw_text.lower()
    my_dict={}
    for p in list(string.punctuation):
        my_dict[p]=' '
    my_dict['\n']=' '
    my_dict["'"]=''
    
    trantab = str.maketrans(my_dict)
    raw_text = raw_text.translate(trantab)
    
    tokens=nltk.word_tokenize(raw_text)
    if add_to_words:
        for t in tokens:
            all_original_words[t]=True
        
    return tokens

def prepare_text(raw_text, add_to_words=True):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization

    Parameters
    ----------
    raw_text : str
        The title or plot of a movie

    Returns
    -------
    list
        A list of tokens
    """
    global lemmatizer,stemmer
    
    tokens = tokenize(raw_text, add_to_words)
    tokens = [lemmatizer.lemmatize(lemmatizer.lemmatize(t,'v'),'a') for t in tokens]
    tokens = [stemmer.stem(d) for d in tokens]
    
    return tokens
    # TODO: tokenize, case_folding, stem, lemmatize
    # return ["edvard", "am", "a", "run", "he", "am", "always", "run"]

prepare_text("Edvard was a Runner. He was always running.")


movies_file=[]

def read_csv_file(file_name):
    ans=[]
    csv_file=open(file_name,'r')
    csv_reader = csv.reader(csv_file, delimiter=',')
    flag = False
    for row in csv_reader:
        if flag:
            ans.append([int(row[0]),prepare_text(row[1]),prepare_text(row[2])])
        flag = True
    return ans

def main():
    global movies_file
    movies_file=read_csv_file('movies.csv')
    
    
main()
```

    <h1>
    <b>ذخیره و فشرده‌سازی نمایه (۲۰ نمره)</b>
    </h1>
    در این بخش باید توانایی ذخیره کردن نمایه و بارگذاری مجدد آن را به سامانه اضافه کنید. ذخیره‌سازی به ۳ روش صورت می‌گیرد. بدون فشرده‌سازی، فشرده‌سازی از روش gamma-code و فشرده‌سازی از روش variable-byte. روش‌های فشرده‌سازی باید توسط خودتان پیاده‌سازی شود.
    برای ذخیره نمایه در فایل نیز از JSON  استفاده کنید.<br>
     بخشی از نمره شما در این قسمت به میزان فشرده‌سازی نمایه اختصاص داده شده است. بنابراین پیاده‌سازی بهینه روش‌های فشرده‌سازی مهم است.

```python
import json
from pathlib import Path
from copy import deepcopy

base=35

def chr2(x):
    if x==ord('\\'):
        x=125
    return chr(x)

def variable_len_enc(x):
    ans=0
    cnt=0
    while x>0 or cnt==0:
        ans<<=1
        ans|=1
        ans<<=7
        ans|=x&((1<<7)-1)
        #print(ans,x,cnt)
        x>>=7
        cnt+=1
    ans^=1<<7
    ans2=0
    while ans>0:
        ans2<<=8
        ans2|=ans&((1<<8)-1)
        ans>>=8
        
    return ans2,cnt

def prepare_list_enc(plist):
    lst=[plist[0]]+plist[1]
    for i in range(len(lst)-1,1,-1):
        if lst[i]>lst[i-1]:
            lst[i]-=lst[i-1]
        else:
            print('Error in variable len enc function!')
    return lst

def variable_len_enc_list(plist):
    global base
    lst=prepare_list_enc(plist)
    
    ans=0
    cnt=0
    out=""
    for x in lst:
        enc=variable_len_enc(x)
        cnt+=8*enc[1]
        ans<<=8*enc[1]
        ans|=enc[0]
        while cnt>=6:
            cnt-=6
            out+=chr2(base+(ans>>cnt))
            ans&=(1<<cnt)-1
    if cnt>0:
        out+=chr2(base+(ans<<(6-cnt)))
    
    return out
    # har 6 bit ra daroon yek char magozaram! chon tedadi az char ha
    # non-printable hastan va be tor khoobi dar json zakhire nemishavand!

def gamma_enc(x):
    b=bin(x)
    s='1'*(len(b)-3)+'0'+b[3:]
    return int(s,2),len(s)



def gamma_enc_list(plist):
    global base
    lst=prepare_list_enc(plist)
    
    ans=0
    cnt=0
    out=""
    for x in lst:
        enc=gamma_enc(x+1)
        cnt+=enc[1]
        ans<<=enc[1]
        ans|=enc[0]
        while cnt>=6:
            cnt-=6
            out+=chr2(base+(ans>>cnt))
            ans&=(1<<cnt)-1
    if cnt>0:
        out+=chr2(base+(ans<<(6-cnt))+(1<<(6-cnt))-1)
    
    return out
    # har 6 bit ra daroon yek char magozaram! chon tedadi az char ha
    # non-printable hastan va be tor khoobi dar json zakhire nemishavand!
    
    

def store_index(path, compression_type):
    """Stores the index in a file

    Parameters
    ----------
    path : str
        Path to store the file

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte

    Returns
    -------
    int
        The size of the stored file
    """
    global positional_index
    try:
        f = open(path, "x")
    except:
        pass
    f=open(path,'w')

    if compression_type=='no-compression':
        f.write(json.dumps(positional_index))
        
    elif compression_type=='variable-byte':
        new_positional_index={}
        for token in positional_index:
            new_positional_index[token]=[[],[]]
            for i in range(2):
                for lst in positional_index[token][i]:
                    new_positional_index[token][i].append(variable_len_enc_list(lst))
            
        f.write(json.dumps(new_positional_index))
        
    elif compression_type=='gamma-code':
        new_positional_index={}
        for token in positional_index:
            new_positional_index[token]=[[],[]]
            for i in range(2):
                for lst in positional_index[token][i]:
                    new_positional_index[token][i].append(gamma_enc_list(lst))
            
        f.write(json.dumps(new_positional_index))
    
    size_of_file = 1024
    
    f.close()
    
    return Path(path).stat().st_size

#store_index("path/to/file.json", 'gamma-code')
```

```python
import json 

base=35

def ord2(c):
    if ord(c)==125:
        return ord('\\')
    return ord(c)

def ok_lst(lst, offset):
    if len(lst)<2:
        print('Error in ok_lst!', lst)
        
    for i in range(len(lst)):
        lst[i]+=offset
        
    ans=[lst[1]]
    for x in lst[2:]:
        ans.append(ans[-1]+x)
    return [lst[0],ans]

def variable_len_dec_list(s):
    global base
    ans=0
    cnt=0
    lst=[]
    for c in s:
        x=ord2(c)-base
        cnt+=6
        ans<<=6
        ans|=x
        while cnt>=8:
            cnt-=8
            z=ans>>cnt
            if (z&(1<<7))==0:
                lst.append(0)
            lst[-1]<<=7
            lst[-1]|=z&((1<<7)-1)
            ans&=(1<<cnt)-1
    
    return ok_lst(lst,0)

def gamma_dec_list(s):
    global base
    ans=0
    cnt=0
    lst=[]
    mod=0
    
    for c in s:
        x=ord2(c)-base
        x=bin(x)[2:]
        x='0'*(6-len(x))+x
        #print(x)
        i=0
        while i<len(x):
            if mod==0:
                while i<len(x):
                    cnt+=1
                    i+=1
                    if x[i-1]=='0':
                        mod=1
                        ans=1
                        cnt-=1
                        break
            if mod==1:
                while cnt>0 and i<len(x):
                    cnt-=1
                    ans<<=1
                    if x[i]=='1':
                        ans|=1
                    i+=1

                if cnt==0:
                    lst.append(ans)
                    mod=0
    
    return ok_lst(lst,-1)

def load_index(path, compression_type):
    """Loads the index from a file

    Parameters
    ----------
    path : str
        Path of the file to load from

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte
    """
    global positional_index
    # TODO: load and decompress positional_index
    f=open(path,'r')
    if compression_type=='no-compression':
        positional_index2=json.loads(f.read())
    elif compression_type=='variable-byte':
        positional_index2=json.loads(f.read())
        for token in positional_index2:
            for i in range(2):
                lst=[]
                for s in positional_index2[token][i]:
                    lst.append(variable_len_dec_list(s))
                positional_index2[token][i]=lst
    elif compression_type=='gamma-code':
        positional_index2=json.loads(f.read())
        for token in positional_index2:
            for i in range(2):
                lst=[]
                for s in positional_index2[token][i]:
                    lst.append(gamma_dec_list(s))
                positional_index2[token][i]=lst
    
    
    #print(positional_index==positional_index2)
    positional_index=positional_index2
    f.close()
    return

#load_index("path/to/file.json", 'gamma-code')
```

```python
# baraye edit distance az site 'geeks for geeks' estefade kardam

def editDistDP(str1, str2):
    m=len(str1)
    n=len(str2)
    # Create a table to store results of subproblems
    dp = [[0 for x in range(n + 1)] for x in range(m + 1)]
 
    # Fill d[][] in bottom up manner
    for i in range(m + 1):
        for j in range(n + 1):
 
            # If first string is empty, only option is to
            # insert all characters of second string
            if i == 0:
                dp[i][j] = j    # Min. operations = j
 
            # If second string is empty, only option is to
            # remove all characters of second string
            elif j == 0:
                dp[i][j] = i    # Min. operations = i
 
            # If last characters are same, ignore last char
            # and recur for remaining string
            elif str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
 
            # If last character are different, consider all
            # possibilities and find minimum
            else:
                dp[i][j] = 1 + min(dp[i][j-1],        # Insert
                                   dp[i-1][j],        # Remove
                                   dp[i-1][j-1])    # Replace
 
    return dp[m][n]

bigram={}
def get_biwords(word):
    word='$'+word+'$'
    biwords=sorted([word[i]+word[i+1] for i in range(len(word)-1)])
    out=[]
    for x in biwords:
        if len(out)==0 or out[-1]!=x:
            out.append(x)
    return out

def create_bigram_index():
    """Creates the bigram index for spell correction"""
    global all_original_words,bigram
    
    
    bigram = {}
    
    lst=sorted(all_original_words)
    all_original_words={}
    for x in lst:
        all_original_words[x]=True
        
    all_chars=['$']+[chr(x) for x in range(ord('a'),ord('z')+1)]
    all_pairs=[x+y for x in all_chars for y in all_chars]
    
    
    for x in all_original_words:
        flag=False
        for c in x:
            if ord(c)<ord('a') or ord(c)>ord('z'):
                flag=True
        if flag:
            continue
        for biword in get_biwords(x):
            if biword not in bigram:
                bigram[biword]=[]
            bigram[biword].append(x)
    #print(cnt/len(all_original_words))
    return bigram

def create_bigram_main():
    create_bigram_index()
create_bigram_main()
```

```python
def jaccard_value(x,y):
    disjoint=0
    bx=get_biwords(x)
    by=get_biwords(y)
    for t in bx:
        if t in by:
            disjoint+=1
    union=len(bx)+len(by)-disjoint
    if union==0:
        return 0.0
    return disjoint/union
    
def get_corrected_word(token):
    global all_original_words,bigram
    NUMBER_OF_BEST_JACC=5
    jacc_value={}
    
    if token in all_original_words:
        return token
    
    for b in get_biwords(token):
        if b not in bigram:
            continue
        for t in bigram[b]:
            if t in jacc_value:
                continue
            jacc_value[t]=jaccard_value(token,t)
    
    top_jaccard=[x[0] for x in sorted(jacc_value.items(),key=lambda x:x[1])[-NUMBER_OF_BEST_JACC:]][::-1]
    
    if len(top_jaccard)==0:
        return token
    best=top_jaccard[0]
    best_edit_dist=editDistDP(token,best)
    
    for x in top_jaccard:
        tmp=editDistDP(token,x)
        #print(x,tmp)
        if tmp<best_edit_dist:
            best=x
            best_edit_dist=tmp
    return best

def get_corrected_text(raw_text):
    """Corrects the query

    Parameters
    ----------
    raw_text : str
        Input text that could be a title or a plot

    Returns
    -------
    str
        The corrected text
    """

    # TODO: correct the query using bigram, jaccard, and edit-distance
    tokens=tokenize(raw_text,False)
    #print(tokens)
    ans=[]
    for t in tokens:
        ans.append(get_corrected_word(t))
        
    return " ".join(ans)

get_corrected_text("the adevntures of herlock holmes")
```

```python
def tpfnfp(actual, predicted):
    ans=0
    for x in predicted:
        if x in actual:
            ans+=1
            
    return ans,len(actual)-ans,len(predicted)-ans


def precision(actual, predicted):
    """Calculate precision

    Parameters
    ----------
    actual : list
        [1st_doc_id, 2nd_doc_id, ...]

    predicted : list
        [1st_doc_id, 2nd_doc_id, ...]

    Returns
    -------
    float
        Precision score
    """

    # TODO: implement precision
    tmp = tpfnfp(actual,predicted)
    return tmp[0]/(tmp[0]+tmp[2])

precision([1,2], [1, 4, 5])
```

```python
def recall(actual, predicted):
    """Calculate recall

    Parameters
    ----------
    actual : list
        [1st_doc_id, 2nd_doc_id, ...]

    predicted : list
        [1st_doc_id, 2nd_doc_id, ...]
    
    Returns
    -------
    float
        Recall score
    """

    # TODO: implement recall
    tmp = tpfnfp(actual,predicted)
    return tmp[0]/(tmp[0]+tmp[1])

recall([1,2], [1, 4, 5])
```

```python
def f1_score(actual, predicted):
    """Calculate f1_score

    Parameters
    ----------
    actual : list
        [1st_doc_id, 2nd_doc_id, ...]

    predicted : list
        [1st_doc_id, 2nd_doc_id, ...]
    
    Returns
    -------
    float
        f1_score
    """

    # TODO: implement f1_score
    p=precision(actual,predicted)
    r=recall(actual,predicted)
    
    f1_score = 0.4
    if p+r==0.0:
        return 0.0
    return 2*p*r/(p+r)

f1_score([1,2], [1, 4, 5])
```

```python
def map_score(actual, predicted):
    """Calculate map_score

    Parameters
    ----------
    actual : list
        [1st_doc_id, 2nd_doc_id, ...]

    predicted : list
        [1st_doc_id, 2nd_doc_id, ...]
    
    Returns
    -------
    float
        map_score
    """

    # TODO: implement map_score
    map_score = 0.0
    curr = 0
    
    for i in range(len(predicted)):
        x=predicted[i]
        if x in actual:
            curr+=1
            map_score+=curr/(i+1)
    if len(actual)==0:
        return 0.0
    return map_score/len(actual)

map_score([1,2,3,4,5,6], [1, 7,2, 3,4,5,8,9,10,6])
```

```python
from math import log

def ndcg_score(actual, predicted):
    """Calculate ndcg_score

    Parameters
    ----------
    actual : list
        [1st_doc_id, 2nd_doc_id, ...]

    predicted : list
        [1st_doc_id, 2nd_doc_id, ...]
    
    Returns
    -------
    float
        ndcg_score
    """

    # TODO: implement recall
    ndcg_score = 0.4
    
    dcggt=0.0
    for i in range(min(len(predicted),len(actual))):
        if i==0:
            dcggt+=1.0
        else:
            dcggt+=1.0/log(i+1,2)
    
    dcgf=0
    for i in range(len(predicted)):
        x=predicted[i]
        if x not in actual:
            continue
        if i==0:
            dcgf+=1
        else:
            dcgf+=1/log(i+1,2)
    
    if dcggt==0:
        return 0.0
    return dcgf/dcggt

ndcg_score([1,2], [1, 4, 5])
```
